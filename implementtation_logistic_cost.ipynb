```python
import numpy as np;
import matplotlib.pyplot as plt;
from lab_utils_common import plot_data, sigmoid, dlc;
```

```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])  #(m,n)
y_train = np.array([0, 0, 0, 1, 1, 1])
```

```python
fig,ax = plt.subplots(1,1,figsize=(4,4));
plot_data(X_train, y_train,ax);
ax.axis([0,4,0,3.5]);
ax.set_xlabel('$x0$', fontsize=12);
ax.set_ylabel('$x1$', fontsize=12);
plt.show();
```

```python
def compute_cost_logistic(x,y,w,b):
    """cost function calculator for logistic regression"""
    m = x.shape[0];
    cost = 0.0;
    
    for i in range(m):
        z_i = np.dot(x[i],w)+b;
        f_wb_i = sigmoid(z_i)
        cost+= -(y[i])*np.log(f_wb_i)-(1-y[i])*np.log(1-f_wb_i);        
        
    cost = cost/m;
    return cost;
    
```

```python
w_tmp = np.array([1,1])
b_tmp = -3
print(compute_cost_logistic(X_train, y_train, w_tmp, b_tmp))
```

```python
x0 = np.arange(0,6);
x1 = 3-x0;
x2 = 4-x0;

fig, ax = plt.subplots(1,1,figsize=(4,4));
plot_data(X_train, y_train, ax);
ax.axis([0,4,0,4]);
ax.plot(x0,x1,c="b");
ax.plot(x0, x2, c="r");
ax.set_ylabel('$x_1$', fontsize=12)
ax.set_xlabel('$x_0$', fontsize=12)
plt.legend(loc="upper right")
plt.title("Decision Boundary")
plt.show();
```

```python
w_arr1 = np.array([1,1]);
b_1 = -3
w_arr2 = np.array([1,1]);
b_2 = -4
print("for b=-3 : cost = ",compute_cost_logistic(X_train, y_train, w_arr1, b_1))
print("for b=-4 : cost = ",compute_cost_logistic(X_train, y_train, w_arr2, b_2))
```
